Making Swiss Roll and implementing Principal component analysis

```python
#implement pca 
#plot swiss roll dataset
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
data=make_swiss_roll(n_samples=3000)
X=data[0]
c=data[1]
ax=plt.axes(projection ="3d")
ax.scatter3D(X[:,0],X[:,1],X[:,2],c=c)
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('x_3')
plt.show()
```

```python
#solve for two pc vectors using the numpy package 
import numpy as np
X_centered=X-X.mean(axis=0)
U,s,Vt=np.linalg.svd(X_centered)
c1=Vt.T[:,0]#eigen vector 1
c2=Vt.T[:,1]#eigen vector 2
W2=Vt.T[:,:2]#two wigen vectors in the column
X2D=X_centered.dot(W2)#dot product of X-centered and w2
plt.scatter(X2D[:,0],X2D[:,1])
plt.show()
```

PCA USING SECOND METHOD

```python
#pca using sklearn package
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X2D=pca.fit_transform(X)
ax=plt.axes()
ax.scatter(X2D[:,0],X2D[:,1])
ax.set_xlabel("PC_1")
ax.set_ylabel("PC_1")
plt.show()
```

```python
#reconstruct training set
X_recovered=pca.inverse_transform(X2D)
#calculate the variance preserved by principal component
Variance_pcs=pca.explained_variance_ratio_
total_variance_pcs=Variance_pcs.sum()
total_variance_pcs
```
