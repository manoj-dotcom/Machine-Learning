```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
```

```python
import pandas as pd
df=pd.DataFrame(data=data.data,columns=data.feature_names)
type(df)
```

```python
df.head(5)
```

```python
df.info
```

```python
df.describe()
```

```python
df.describe().iloc[[1,2,3,7],]
```

```python
import matplotlib.pyplot as plt
ax=plt.axes()
ax.scatter(df.iloc[:,0],df.iloc[:,1],c=data.target)
ax.set_xlabel(df.columns[0]+'(x1)')
ax.set_ylabel(df.columns[1]+'(x2)')
plt.show()
```

```python
ax=plt.axes(projection="3d")
ax.scatter3D(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],c=data.target)
ax.set_xlabel(df.columns[0]+'(x1)')
ax.set_ylabel(df.columns[1]+'(x2)')
ax.set_zlabel(df.columns[2]+'(x3)')
plt.show()
```
